```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Life Expectancy Data.csv')

# to confirm the data is loaded in properly
df.head()
```

```python
print(f"Shape (rows, columns): {df.shape}")
print(f"Size (rows * columns): {df.size}")

df.isnull().sum()
```

```python
df = df.drop(['Population', 'Hepatitis B', 'GDP', 'percentage expenditure', 'Income composition of resources', 'Schooling'], axis=1)
```

```python
df['Life expectancy '] = df['Life expectancy '].fillna(df['Life expectancy '].median())
df['Adult Mortality'] = df['Adult Mortality'].fillna(df['Adult Mortality'].median())
df[' BMI '] = df[' BMI '].fillna(df[' BMI '].median())
df['Polio'] = df['Polio'].fillna(df['Polio'].median())
df['Diphtheria '] = df['Diphtheria '].fillna(df['Diphtheria '].median())
df[' thinness  1-19 years'] = df[' thinness  1-19 years'].fillna(df[' thinness  1-19 years'].median())
df[' thinness 5-9 years'] = df[' thinness 5-9 years'].fillna(df[' thinness 5-9 years'].median())
df['Alcohol'] = df['Alcohol'].fillna(df['Alcohol'].median())
```

```python
# Dropped all rows with missing values
# df.dropna(inplace=True)
df.isnull().sum()
```

```python
print(df.columns)
```

```python
plt.scatter(df[' thinness  1-19 years'], df['Life expectancy '])
plt.xlabel('Thinness')
plt.ylabel('Life Expectancy')
plt.title('Thinness vs Life Expectancy')
plt.ticklabel_format(style='plain')
plt.show()

plt.scatter(df[' BMI '], df['Life expectancy '])
plt.xlabel('BMI')
plt.ylabel('Life Expectancy')
plt.title('BMI vs Life Expectancy')
plt.ticklabel_format(style='plain')
plt.show()

plt.scatter(df['Adult Mortality'], df['Life expectancy '])
plt.xlabel('Adult Mortality')
plt.ylabel('Life Expectancy')
plt.title('Adult Mortality vs Life Expectancy')
plt.ticklabel_format(style='plain')
plt.show()

plt.scatter(df['Alcohol'], df['Life expectancy '])
plt.xlabel('Alcohol')
plt.ylabel('Life Expectancy')
plt.title('Alcohol vs Life Expectancy')
plt.ticklabel_format(style='plain')
plt.show()

plt.scatter(df['Measles '], df['Life expectancy '])
plt.xlabel('Measles')
plt.ylabel('Life Expectancy')
plt.title('Measles vs Life Expectancy')
plt.ticklabel_format(style='plain')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import seaborn as sns
```

```python
x = df[['Physical Activity Level', 'Stress Level', 'Sleep Duration']]
y = df['Life expectancy']

sleepQuality_model = LinearRegression()
sleepQuality_model.fit(x, y)

coefficients = sleepQuality_model.coef_
intercept = sleepQuality_model.intercept_

# Printing the intercept and coefficients
print("Intercept:", intercept)
print("Coefficients:")
for i, coef in enumerate(coefficients):
    print(f"{i + 1}: {coef}")
```
